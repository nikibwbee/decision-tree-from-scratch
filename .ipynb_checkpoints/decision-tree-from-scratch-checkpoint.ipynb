```python
import pandas as pd
df= pd.read_csv("tennis.csv")
df.head()
```

```python
x= df.drop(columns="PlayTennis")
y= df["PlayTennis"]
```

```python
import numpy as np

def entropy(y):
    #find unique values and their counts
    values, counts =np.unique(y, return_counts =True)
    
    probs=counts / counts.sum()

    ent=0
    for p in probs:
        ent -= p * np.log2(p)

    return ent 
    
total_entropy= entropy(y)
print("Total entropy:",total_entropy)
```

```python
 # for predicting gain for the feature

def info_gain(df, feature, target="PlayTennis"):
    
    total_entropy =entropy(df[target])

    values = df[feature].unique()

    weighted_entropy =0

    for value in values:
        
        subset = df[df[feature]==value]

        subset_entropy =entropy(subset[target])
        weight =len(subset) /len(df)
        weighted_entropy +=weight * subset_entropy

        print(f"Value: {value}, Subset entropy: {subset_entropy}, Weight: {weight}")

    gain = total_entropy - weighted_entropy
    print(f"Info gain by splitting on {feature}: {gain}\n")
    return gain

```

```python
# resursive desicion tree builder

def build_tree(df, features, target, depth=0, max_depth=3):
    
    y=df[target]

    #checking different values in this subset of the target column, if only 1 stop=leaf node
    if len(y.unique()) == 1:
        return y.iloc[0]

    #checking if no features left or max dept reached, return most repeated value
    if len(features) == 0 or depth == max_depth:
        return y.mode()[0]

  

    gains = {}
    
    for feature in features:
        gain = info_gain(df,feature,target)
        gains[feature] = gain
    
    #finding the feature with the highest gain
    best_feature = max(gains, key=gains.get)

    
    #create the tree node for the best feature ~ recursive
    tree={best_feature: {}}
    
    #make branches for each value
    
    for value in df[best_feature].unique():
        subset = df[df[best_feature] == value].copy()
        sub_features = [f for f in features if f != best_feature]
        
        tree[best_feature][value] = build_tree(subset, sub_features, target, depth + 1, max_depth)
        
    return tree
```
